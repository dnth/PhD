```python
import pandas
from pandas import read_csv
```

### Load the dataset

```python
dataframe_0 = read_csv('data_prep/DST_0C.csv', header=0, index_col=0)
dataframe_25 = read_csv('data_prep/DST_25C.csv', header=0, index_col=0)
dataframe_45 = read_csv('data_prep/DST_45C.csv', header=0, index_col=0)
dataframe_DST = dataframe_0.append(dataframe_25)
dataframe_DST = dataframe_DST.append(dataframe_45)

dataframe_DST = dataframe_DST.reset_index(drop=True)
```

```python
dataframe_0 = read_csv('data_prep/FUDS_0C.csv', header=0, index_col=0)
dataframe_25 = read_csv('data_prep/FUDS_25C.csv', header=0, index_col=0)
dataframe_45 = read_csv('data_prep/FUDS_45C.csv', header=0, index_col=0)
dataframe_FUDS = dataframe_0.append(dataframe_25)
dataframe_FUDS = dataframe_FUDS.append(dataframe_45)

dataframe_FUDS = dataframe_FUDS.reset_index(drop=True)
```

```python
dataframe_0 = read_csv('data_prep/BJDST_0C.csv', header=0, index_col=0)
dataframe_25 = read_csv('data_prep/BJDST_25C.csv', header=0, index_col=0)
dataframe_45 = read_csv('data_prep/BJDST_45C.csv', header=0, index_col=0)
dataframe_BJDST = dataframe_0.append(dataframe_25)
dataframe_BJDST = dataframe_BJDST.append(dataframe_45)

dataframe_BJDST = dataframe_BJDST.reset_index(drop=True)
```

```python
dataframe_0 = read_csv('data_prep/US06_0C.csv', header=0, index_col=0)
dataframe_25 = read_csv('data_prep/US06_25C.csv', header=0, index_col=0)
dataframe_45 = read_csv('data_prep/US06_45C.csv', header=0, index_col=0)
dataframe_US06 = dataframe_0.append(dataframe_25)
dataframe_US06 = dataframe_US06.append(dataframe_45)

dataframe_US06 = dataframe_US06.reset_index(drop=True)
```

### Visualize Dataset

```python
%matplotlib inline
import matplotlib.style as style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
ax = dataframe_DST.plot()
ax.set_title("DST")
ax.set_ylabel("Values")
ax.set_xlabel("Timestep")
# plt.savefig(fname='dataframe_DST', transparent=True, dpi=100, bbox_inches='tight')
```

```python
ax = dataframe_FUDS.plot()
ax.set_title("FUDS")
ax.set_ylabel("Values")
ax.set_xlabel("Timestep")
# plt.savefig(fname='dataframe_DST', transparent=True, dpi=100, bbox_inches='tight')
```

```python
ax = dataframe_BJDST.plot()
ax.set_title("BJDST")
ax.set_ylabel("Values")
ax.set_xlabel("Timestep")
# plt.savefig(fname='dataframe_DST', transparent=True, dpi=100, bbox_inches='tight')
```

```python
ax = dataframe_US06.plot()
ax.set_title("US06")
ax.set_ylabel("Values")
ax.set_xlabel("Timestep")
# plt.savefig(fname='dataframe_DST', transparent=True, dpi=100, bbox_inches='tight')
```

### Combine all dataset

```python
dataframe_all = dataframe_DST.append(dataframe_FUDS)
dataframe_all = dataframe_all.append(dataframe_BJDST)
dataframe_all = dataframe_all.append(dataframe_US06)

dataframe_all = dataframe_all.reset_index(drop=True)
```

```python
ax = dataframe_all.plot()
ax.set_title("All")
ax.set_ylabel("Values")
ax.set_xlabel("Timestep")
```

```python
dataframe_even_index = dataframe_all.iloc[::2]
ax = dataframe_even_index.plot()
ax.set_title("Even Index")
ax.set_ylabel("Values")
ax.set_xlabel("Timestep")
```

```python
dataframe_odd_index = dataframe_all.iloc[1::2]
ax = dataframe_odd_index.plot()
ax.set_title("Odd Index")
ax.set_ylabel("Values")
ax.set_xlabel("Timestep")
```

```python
dataframe_testset = dataframe_odd_index.iloc[0::2]
ax = dataframe_testset.plot()
ax.set_title("Test Set")
ax.set_ylabel("Values")
ax.set_xlabel("Timestep")
len(dataframe_testset)
dataframe_testset.to_csv('test_set.csv')
```
